```python
class BrowserHistory:

    def __init__(self, homepage: str):
        self.list1   = []
        self.current = -1

        self.list1.append(homepage)
        self.current+=1

    def visit(self, url: str) -> None:
        if (self.current == len(self.list1) - 1):
            self.list1.append(url)
            self.current+=1

            print("         Primary add")
            print("LIST    : ",self.list1)
            print("CURRENT : ",self.current)

    
        else:
            while (self.current < len(self.list1) -1 ):  ## Visits url from the current page. It clears up all the forward history.
                print("     WHILE LOOP")
                print("LIST    : ",self.list1)
                print("CURRENT : ",self.current)
                self.list1.pop()

            self.list1.append(url)
            self.current+=1
            print("     AFTER WHILE LOOP")
            print("LIST    : ",self.list1)
            print("CURRENT : ",self.current)

    def back(self, steps: int) -> str:
        if self.current - steps >= 0:
            self.current = self.current - steps
            
            print("     BACK")
            print("LIST    : ",self.list1)
            print("CURRENT : ",self.current)

            return self.list1[self.current]
        else:
            self.current-=1
            return self.list1[self.current]
        
    def forward(self, steps: int) -> str:
        if (self.current + steps < len(self.list1) - 1):
            self.current = self.current + steps
            
            print("     FORWARD")
            print("LIST    : ",self.list1)
            print("CURRENT : ",self.current)
            return self.list1[self.current]
        else:

            return self.list1[self.current]
        


# Your BrowserHistory object will be instantiated and called as such:
# obj = BrowserHistory(homepage)
# obj.visit(url)
# param_2 = obj.back(steps)
# param_3 = obj.forward(steps)
```
