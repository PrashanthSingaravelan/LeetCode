```python
class CustomStack:

    def __init__(self, maxSize: int):
        self.list1 = []
        self.maxSize = maxSize

    def push(self, x: int) -> None:
        if len(self.list1) < self.maxSize:
            self.list1.append(x)

    def pop(self) -> int:

        if len(self.list1) > 0 :
            element = self.list1.pop()
            return element
        else:
            return -1
        
    def increment(self, k: int, val: int) -> None:
        if k < len(self.list1):
            for i in range(k):
                self.list1[i] = self.list1[i] + val

        else:
            self.list1 = [(i+val) for i in self.list1]
```

```python
##Your CustomStack object will be instantiated and called as such:
maxSize = 3
obj = CustomStack(maxSize)
x = 10
obj.push(x)
param_2 = obj.pop()
k=2
val=100
obj.increment(k,val)
```
