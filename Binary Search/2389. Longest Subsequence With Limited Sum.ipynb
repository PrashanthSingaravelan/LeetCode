```python
prefix_list = []

def prefix_method(list1):
    prefix_list.append(list1[0])
    for i in range(1, len(list1)):
        prefix_list.append(prefix_list[i-1] + list1[i])

def binary_search(low, high, element):
    middle = (low + high)/2
    middle = int(middle)
    
    if (low<=high):
        if (element == prefix_list[middle]):
            return element

        elif (element <= prefix_list[middle]):
            return binary_search(low, middle - 1, element)

        elif (element >= prefix_list[middle]):
            return binary_search(middle+1, high, element)
    
    return (prefix_list[middle])

nums = [4,5,2,1]
queries = [3,10,21]

nums.sort()
prefix_method(nums)

element_ans = []

for query_index in range(len(queries)):
    element = binary_search(0,len(prefix_list) -1, queries[query_index]);
    element_ans.append(element)
    
ans = []

if (len(element_ans) == 1):
    ans.append(prefix_list.index(element_ans[0]))
else:
    for i in element_ans:
        ans.append(prefix_list.index(i) + 1)

print(ans)
```
